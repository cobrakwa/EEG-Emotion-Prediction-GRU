# Importing Libraries

```python
# Working with data
import numpy as np
import pandas as pd

# Data Visualisation
import matplotlib.pyplot as plt
import seaborn as sns

# Preprocessing
from sklearn.model_selection import train_test_split

# Modelling
import tensorflow as tf

# Results analysis
from sklearn.metrics import confusion_matrix, classification_report
```

# Dataset

```python
data = pd.read_csv('../input/eeg-brainwave-dataset-feeling-emotions/emotions.csv')
data
```

##  Visualising Data

```python
# looking at first row
sample_data = data.loc[0,'fft_0_b':'fft_749_b']

plt.figure(figsize=(16, 10))
plt.plot(range(len(sample_data)), sample_data)
plt.title("Features fft_0_b through fft_749_b")
plt.show()
```

## Encoding Labels

```python
# See label classes
data['label'].value_counts()
```

```python
# Create a label mapping
label_mapping = {"NEGATIVE": 0, "NEUTRAL": 1, "POSITIVE": 2}
```

# Preprocessing

```python
# create fucntion to split data
def preprocess_inputs(df):
    df = df.copy()
    
    # encode label classes
    df['label'] = df['label'].replace(label_mapping)
    
    y = df['label'].copy()
    X = df.drop('label', axis=1).copy()
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=123)
    return X_train, X_test, y_train, y_test

# split data
X_train, X_test, y_train, y_test = preprocess_inputs(data)
X_train
    
```

# Model

## 2 Hidden layer network (to comment out)

```python
#create input layer to take vectors of 2548 columns (no. of features)
inputs = tf.keras.Input(shape=(X_train.shape[1],))  

#pass through hidden layers
#note: 64 neurons in each layer is arbitrary to balance overfit & computational cost & efficiency
#note: reLU to learn non-linear relationships betw features, & avoid vanishing gradient problem
x = tf.keras.layers.Dense(64, activation='relu')(inputs)
x = tf.keras.layers.Dense(64, activation='relu')(x)

#note: softmax used for multi-class classification
outputs = tf.keras.layers.Dense(3, activation='softmax')(x) # 3 units since 3 classes

model = tf.keras.Model(inputs=inputs, outputs=outputs)
print(model.summary())
```

# RNN 
RNN allows the keeping track of previous time steps
EEG readings are time series so we need time dependency

```python
inputs = tf.keras.Input(shape=(X_train.shape[1],))  

# expand dim to 3d to be passed into GRU
expand_dims = tf.expand_dims(inputs, axis=2)

gru = tf.keras.layers.GRU(256, return_sequences=True)(expand_dims) # returns 2-dim array
#flatten 2-dim array into 1-dim vector
flatten = tf.keras.layers.Flatten()(gru)

# note: GRU layer accepts 3d vector input, BC each bit of data that comes in is encoded 
# as a vector itself
outputs = tf.keras.layers.Dense(3, activation='softmax')(x)
```

## Configure model for training

```python
# model configuration
model.compile(
    optimizer = 'adam', #popular opti algo for deep learning, needs minimal parameter tuning
    loss='sparse_categorical_crossentropy', #for multi-class catergorical priblems where labels are integers (instead of one-hot encoded vectors)
    metrics=['accuracy'] # good metric to assess model performance, esp w balanced class distribution
)

# training
# history: stores training & validation loss + metric for each epoch
history = model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    batch_size=32,
    epochs=50,
    callbacks=[
#allow model to look at val loss after each epcoh & evaluate if it is improving. If does not improve for 5 epochs, stop trraining and restore weights from best epoch
        tf.keras.callbacks.EarlyStopping( 
            monitor='val_loss',
            patience=5,
            restore_best_weights=True
        )
    ]
)
```

# Results & Model Evaluation

#### Accuracy

```python
accuracy = model.evaluate(X_test, y_test, verbose=0)[1]
print("Test Accuracy: {:.3f}%".format(accuracy * 100))
```

#### Confusion Matrix & Classification Report

```python
# get array of predicted integer classes for each timestep
model.predict(X_test)
y_pred = np.array(list(map(lambda x: np.argmax(x), model.predict(X_test)))) #most probable class
y_pred
cm = confusion_matrix(y_test, y_pred)

# Classification Report
clr = classification_report(y_test, y_pred, target_names = label_mapping.keys())

# Visualise Confusion Matrix
plt.figure(figsize=(8,8))
sns.heatmap(cm, annot=True, vmin=0, fmt='g', cbar=False, cmap='Blues')
# change ticks to word labels
plt.xticks(np.arange(3) + 0.5, label_mapping.keys())
plt.yticks(np.arange(3) + 0.5, label_mapping.keys())
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# Visualise Classification Report
print("Classification Report:\n-----------------------\n", clr)
```
